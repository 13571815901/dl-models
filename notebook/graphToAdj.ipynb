### read JSON

```python
import json
with open('./dataset/membrane.graph.json') as handle:
    graph = json.loads(handle.read())
```

```python
print(type(graph))
print(graph.keys())
print(len(graph))
```

```python
graph = { int(k):v for k,v in graph.items() } # convert string keys to number keys
```

```python
print(graph.items())
```

### convert dict graph to adj matrix

```python
import networkx as nx
```

```python
G = nx.from_dict_of_lists(graph)
adj = nx.adjacency_matrix(nx.from_dict_of_lists(graph))
```

```python
print(type(adj))
print(adj.shape)
print(type(G))
print(G.nodes)
```

```python
adj_d = adj.toarray()
```
